```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Entering data
data = pd.DataFrame([
    [np.nan, 0.5219, 0.6525, 0.2427],
    [0.5219, np.nan, 0.3853, 0.1281],
    [0.6525, 0.3853, np.nan, 0.1526],
    [0.2427, 0.1281, 0.1526, np.nan]
], index=["Liver", "Lung", "Stomach", "Breast"], columns=["Liver", "Lung", "Stomach", "Breast"])

# Create a heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(data, annot=True, cmap="coolwarm", fmt=".3f", linewidths=0.5, linecolor="gray")

plt.title("Gradient similarity")

plt.show()
```
