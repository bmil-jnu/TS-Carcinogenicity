```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = pd.DataFrame([
    [np.nan, 0.6160, 0.7259, 0.1558],
    [0.6160, np.nan, 0.4898, 0.0812],
    [0.7259, 0.4898, np.nan, -0.0382],
    [0.1558, 0.0812, -0.0382, np.nan]
], index=["Liver", "Lung", "Stomach", "Breast"], columns=["Liver", "Lung", "Stomach", "Breast"])

plt.figure(figsize=(6, 5))
sns.heatmap(data, annot=True, cmap="coolwarm", fmt=".3f", linewidths=0.5, linecolor="gray")

plt.title("Gradient similarity")

plt.show()
```
