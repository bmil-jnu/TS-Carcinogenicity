```python
# CSV 파일에서 SMILES 데이터를 읽어옵니다.
df = pd.read_csv("smiles.csv")

# 저장할 디렉토리 경로
output_dir = "/data/home/dbswn0814/multi_task(1106)/model/DCAMCP/liv/liver SMILES"

# 각 SMILES를 개별적인 .smi 파일로 저장합니다.
for index, row in df.iterrows():
    smiles = row["smiles"]
    name = row["name"] if "name" in row else f"{index + 1}"
    
    # .smi 파일로 저장
    smiles_filename = os.path.join(output_dir, f"{name}.smi")
    with open(smiles_filename, "w") as f:
        f.write(f"{smiles} {name}\n")
    
    print(f"SMILES '{smiles}'이(가) '{smiles_filename}'로 저장되었습니다.")
```
